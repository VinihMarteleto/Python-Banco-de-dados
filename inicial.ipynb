### Case - Cancelamento de Clientes

Você foi contratado por uma empresa com mais de 800 mil clientes para um projeto de Dados. Recentemente a empresa percebeu que da sua base total de clientes, a maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.

Base de dados e arquivos: https://drive.google.com/drive/folders/1uDesZePdkhiraJmiyeZ-w5tfc8XsNYFZ?usp=drive_link

```python
import pandas as pd
#importar a biblioteca pandas para manipulação de dados em Python. A biblioteca é comumente usada para análise de dados e oferece estruturas de dados flexíveis, como DataFrames, que facilitam a manipulação e análise de conjuntos de dados.
tabela = pd.read_csv("cancelamentos.csv")
tabela = tabela.drop(columns="CustomerID")
display(tabela)
```

```python

display(tabela.info()) #exibe informações sobre o DataFrame, como o número de linhas, colunas, tipos de dados e valores nulos.

tabela = tabela.dropna() #remove as linhas que contêm valores nulos do DataFrame "tabela". Isso é útil para garantir que os dados estejam completos e prontos para análise, evitando problemas causados por valores ausentes.

display(tabela.info()) #exibe informações sobre o DataFrame "tabela" após a remoção de valores nulos, mostrando o número de linhas, colunas, tipos de dados e confirmando que não há mais valores nulos presentes.
```

```python
#contar os cancelamentos por motivo
display(tabela["cancelou"].value_counts())

#contar os cancelamentos por motivo em porcentagem
display(tabela["cancelou"].value_counts(normalize=True).map("{:.3%}".format))

```
