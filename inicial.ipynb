### Case - Cancelamento de Clientes

Você foi contratado por uma empresa com mais de 800 mil clientes para um projeto de Dados. Recentemente a empresa percebeu que da sua base total de clientes, a maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.

Base de dados e arquivos: https://drive.google.com/drive/folders/1uDesZePdkhiraJmiyeZ-w5tfc8XsNYFZ?usp=drive_link

```python

import pandas as pd

tabela = pd.read_csv("cancelamentos.csv")

# Passo 2: Visualizar a base de dados
if 'CustomerID' in tabela.columns:
    tabela = tabela.drop(columns='CustomerID')
display(tabela)
# colunas inúteis - informações que não te ajudam, te atrapalham
```

```python
# Passo 4: Análise inicial dos cancelamentos

# quantas pessoas cancelaram e quantas não cancelaram
display(tabela["cancelou"].value_counts())

# em percentual
display(tabela["cancelou"].value_counts(normalize=True))
# display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# clientes do contrato mensal TODOS cancelam
    # ofecer desconto nos planos anuais e trimestrais
# clientes que ligam mais do que 4 vezes para o call center, cancelam
    # criar um processo para resolver o problema do cliente em no máximo 3 ligações
# clientes que atrasaram mais de 20 dias, cancelaram
    # política de resolver atrasos em até 10 dias (equipe financeira)

# normalizar tipos e aplicar filtros de forma segura
tabela['ligacoes_callcenter'] = pd.to_numeric(tabela['ligacoes_callcenter'], errors='coerce')
tabela['dias_atraso'] = pd.to_numeric(tabela['dias_atraso'], errors='coerce')
if 'duracao_contrato' in tabela.columns:
    tabela = tabela[~tabela['duracao_contrato'].str.lower().eq('monthly')]
tabela = tabela[tabela['ligacoes_callcenter']<=4]
tabela = tabela[tabela['dias_atraso']<=20]

display(tabela['cancelou'].value_counts())
# em percentual
display(tabela['cancelou'].value_counts(normalize=True))
```

```python
# Fallback: força renderer estático e mostra qual Python o kernel usa
import sys
import plotly.io as pio
pio.renderers.default = 'png'
print('Kernel Python:', sys.executable)
try:
    import nbformat
    print('nbformat:', nbformat.__version__)
except Exception as e:
    print('nbformat não disponível:', e)
```

```python
# aplicar filtros e gerar resumo + gráfico seguro
t = tabela.copy()
if 'duracao_contrato' in t.columns:
    t = t[~t['duracao_contrato'].str.lower().eq('monthly')]
if 'ligacoes_callcenter' in t.columns:
    t['ligacoes_callcenter'] = pd.to_numeric(t['ligacoes_callcenter'], errors='coerce')
    t = t[t['ligacoes_callcenter'] <= 4]
if 'dias_atraso' in t.columns:
    t['dias_atraso'] = pd.to_numeric(t['dias_atraso'], errors='coerce')
    t = t[t['dias_atraso'] <= 20]

display(t['cancelou'].value_counts(normalize=True))
# gerar gráfico de cancelamentos após filtros
import plotly.express as px
fig = px.histogram(t, x='cancelou', color='cancelou', title='Cancelamentos após filtros')
fig.write_html('figures/cancelou_filtrado.html')
try:
    fig.show()
except Exception:
    try:
        fig.write_image('figures/cancelou_filtrado.png', scale=2)
    except Exception:
        pass
```
